## SSL Pre-training

```python
import pandas as pd
import matplotlib.pyplot as plt

# paths to your csv files
subset_csv = "/Users/rabinatwayana/1_Rabina/3_CDE_III/ICPR_competition/ICPR-Contest-2026/data/krs_output_branch_rabina2/logs/ssl_subset_v3_e50_b256_mem_8k_rm_norm/metrics_20260227_074423/version_0/metrics.csv"
fullset_csv = "/Users/rabinatwayana/1_Rabina/3_CDE_III/ICPR_competition/ICPR-Contest-2026/data/krs_output_branch_rabina2/logs/ssl_full_v4_e50_b256_mem_16k_rm_norm/metrics_20260227_105432/version_0/metrics.csv"

# read csv files
subset_df = pd.read_csv(subset_csv)
fullset_df = pd.read_csv(fullset_csv)

# plot
plt.figure()

plt.plot(subset_df["epoch"], subset_df["train_loss"], label="Subset", marker='o')
plt.plot(fullset_df["epoch"], fullset_df["train_loss"], label="Fullset", marker='o')

# labels and title
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("Training Loss: Subset vs Fullset")
plt.legend()
plt.grid(True)

plt.show()
```

## Train-val loss using full ssl pre-trained weight in downstreaming task

```python
import pandas as pd
import matplotlib.pyplot as plt
#INPUT: output_metrics_path
output_metrics_path="/Users/rabinatwayana/1_Rabina/3_CDE_III/ICPR_competition/ICPR-Contest-2026/output/downstream/fullset/metrics.csv"

# Load the CSV file
metrics_df = pd.read_csv(output_metrics_path)
metrics_df_ = metrics_df.dropna(how='all') # Drop rows where all values are NaN
metrics_df_ = metrics_df_.ffill() # Forward-fill NaN values (carry last valid observation forward)
metrics_df_ = metrics_df_.groupby('epoch').last().reset_index()

# plots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].plot(metrics_df_["epoch"], metrics_df_["train_loss"], label="Train Loss")
axes[0].plot(metrics_df_["epoch"], metrics_df_["val_loss"], label="Validation Loss")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Loss")
axes[0].set_title("Train vs. Validation Loss")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(metrics_df_["epoch"], metrics_df_["train_f1_macro"], label="Train F1 Macro")
axes[1].plot(metrics_df_["epoch"], metrics_df_["val_f1_macro"], label="Validation F1 Macro")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("F1 Macro")
axes[1].set_title("Train vs. Validation F1 Macro")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
